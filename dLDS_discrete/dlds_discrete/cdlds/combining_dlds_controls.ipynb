```python
import models
from models import DLDSwithControl
from datasets import CdLDSDataGenerator
import main_functions as mf
from importlib import reload
import plotly.express as px
import numpy as np
```

```python
model = DLDSwithControl(CdLDSDataGenerator(K=1, fix_point_change=False, eigenvalue_radius=0.98))

# generate toy data
X = model.datasets.generate_data(1000, sigma=0.01)
```

```python
model_mf = mf.train_model_include_D(data=X, max_iter=1000)
```

```python
fig = px.line(model_mf[3], title='Reconstruction Error (with original function)').update_layout(xaxis_title='Iteration', yaxis_title='Error', showlegend=False)
fig.add_shape(type="line", x0=np.argmin(model_mf[3]), y0=np.max(model_mf[3]), x1=np.argmin(model_mf[3]), y1=np.min(model_mf[3]), line=dict(color="Red", width=1))
fig.show()
```
